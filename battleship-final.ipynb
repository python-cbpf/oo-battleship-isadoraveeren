```python
import numpy as np

from pandas import *
```

```python
def Map(N): #Function to build a 10x10 grid that will play as a board
    matrix=[] 
    for i in range(N): 
        row=[] 
        for j in range(N): 
            row.append(0) 
        matrix.append(row) 
    return matrix
```

```python
MyShipsA=Map(10) #map where player A places their ships
MyShotsA=Map(10) #map where player A takes their guesses

MyShipsB=Map(10) #map where player B places their ships
MyShotsB=Map(10) #map where player B takes their guesses

destroyerA=0 #counters for each time player A's ships where hit
cruiserA=0
submarineA=0
battleshipA=0
carrierA=0

destroyerB=0 #counters for each time player B's ships where hit
cruiserB=0
submarineB=0
battleshipB=0
carrierB=0
```

```python
class PlayerA:
    def __init__(self,name):
        self.name = name 
    
    def ShowMyShips(self): #show player A's ships
        self.ShowMyShips = print(DataFrame(MyShipsA))

    def ShowMyShots(self): #show player A's guesses
        self.ShowMyShots = print(DataFrame(MyShotsA))
        
    def PlaceShip(self,ship,x1,y1,x2,y2,x3 = None,y3 = None,x4 = None,y4 = None,x5 = None,y5 = None): 
        #player A places their ships 
        global MyShipsA
        if (ship == "destroyer" and MyShipsA[x1][y1] == 0 and MyShipsA[x2][y2] == 0): #checks that you don't overlap ships
            MyShipsA[x1][y1]=1 #each different ship is marked by a different number on A's board
            MyShipsA[x2][y2]=1
        elif (ship == "cruiser" and MyShipsA[x1][y1] == 0 and MyShipsA[x2][y2] == 0 and MyShipsA[x3][y3] == 0):
            MyShipsA[x1][y1]=2
            MyShipsA[x2][y2]=2
            MyShipsA[x3][y3]=2
        elif (ship == "submarine" and MyShipsA[x1][y1] == 0 and MyShipsA[x2][y2] == 0 and MyShipsA[x3][y3] == 0):
            MyShipsA[x1][y1]=3
            MyShipsA[x2][y2]=3
            MyShipsA[x3][y3]=3
        elif (ship == "battleship" and MyShipsA[x1][y1] == 0 and MyShipsA[x2][y2] == 0 and MyShipsA[x3][y3] == 0 and MyShipsA[x4][y4] == 0):
            MyShipsA[x1][y1]=4
            MyShipsA[x2][y2]=4
            MyShipsA[x3][y3]=4
            MyShipsA[x4][y4]=4
        elif (ship == "carrier" and MyShipsA[x1][y1] == 0 and MyShipsA[x2][y2] == 0 and MyShipsA[x3][y3] == 0 and MyShipsA[x4][y4] == 0 and MyShipsA[x5][y5] == 0):
            MyShipsA[x1][y1]=5
            MyShipsA[x2][y2]=5
            MyShipsA[x3][y3]=5
            MyShipsA[x4][y4]=5
            MyShipsA[x5][y5]=5
        else:
            print ("Are you sure you picked the right ship? Make sure they don't overlap!") #message in case ships overlap or have wrong size
        
        print(DataFrame(MyShipsA))

        
    def Shoot(self,x,y): #player A takes their guesses
        global MyShotsA
        global MyShipsB
        global destroyerB
        global cruiserB
        global submarineB
        global battleshipB
        global carrierB
        
        if MyShotsA[x][y] != 0: #checks that its a brand new guess
            print ("you already hit this spot! choose another target")
            
        else:
        
            if MyShipsB[x][y] == 0: #if player B placed no ship on (x,y), player A marks an "x" on their guess board
                print ("Splash!")
                MyShotsA[x][y]="x"

            if MyShipsB[x][y] == 1: #if player B placed a destroyer on (x,y), player A marks the spot with a "1".
                print ("you hit my destroyer!") #each different boat hit is marked with a different number
                MyShotsA[x][y]=1
                destroyerB=destroyerB+1
                if destroyerB == 2:
                    print ("you sunk my destroyer!") #when the whole ship is hit, it sinks

            if MyShipsB[x][y] == 2:
                print ("you hit my cruiser!")
                MyShotsA[x][y]=2
                cruiserB=cruiserB+1
                if cruiserB == 3:
                    print ("you sunk my cruiser!")

            if MyShipsB[x][y] == 3:
                print ("you hit my submarine!")
                MyShotsA[x][y]=3
                submarineB=submarineB+1
                if submarineB == 3:
                    print ("you sunk my submarine!")

            if MyShipsB[x][y] == 4:
                print ("you hit my battleship!")
                MyShotsA[x][y]=4
                battleshipB=battleshipB+1
                if battleshipB == 4:
                    print ("you sunk my battleship!")

            if MyShipsB[x][y] == 5:
                print ("you hit my carrier!")
                MyShotsA[x][y]=5
                carrierB=carrierB+1
                if carrierB == 5:
                    print ("you sunk my carrier!")

            if (destroyerB == 2 and cruiserB == 3 and submarineB == 3 and battleshipB == 4 and carrierB == 5):
                print ("you sinked all my ships! GAME OVER")
        
        print(DataFrame(MyShotsA))
```
